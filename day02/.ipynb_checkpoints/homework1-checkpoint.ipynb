```python
import subprocess
import matplotlib.pyplot as plt
%matplotlib inline
N=100
y=[]
for i in range(1,N+1):
    proc = subprocess.run(["python3","homework1.py",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8")))
    #print(proc.stderr.decode("utf8"))
x=list(range(1,N+1))
plt.plot(x,y)
plt.show()
```

```python
y=[]
N=500
proc=subprocess.run(["gcc","-o","homework1","homework1.c"],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
for i in range(1,N+1,10):
    proc = subprocess.run(["./homework1",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8")))
    #print(proc.stderr.decode("utf8"))
x=list(range(1,N+1,10))
plt.plot(x,y)
plt.show()
```

 # Haskell
 なぜか速度が速すぎる
 測定するまでもなく速い(N=1000000とかでも終わるので)
N=10,100,1000でC=ABがちゃんと行列計算できていると思うので、どうしてなのか不思議

```python
y=[]
N=1000000
proc=subprocess.run(["ghc","homework1.hs"],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
for i in range(1,N+1,10000):
    proc = subprocess.run(["./homework1",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8").strip()[:-1]))
    #print(proc.stderr.decode("utf8"))
print(y)
x=list(range(1,N+1,10000))
plt.plot(x,y)
plt.show()
```
