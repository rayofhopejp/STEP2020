# Python

```python
import subprocess
import matplotlib.pyplot as plt
%matplotlib inline
N=50
y=[]
for i in range(1,N+1,1):
    proc = subprocess.run(["python3","homework1.py",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8").split()[0]))
    #print(proc.stdout.decode("utf8"))
x=list(range(1,N+1,1))
plt.plot(x,y)
plt.show()
```

# C
Pythonより速い。
Pythonより速いのは、インタープリタ言語ではないから、かつ静的な型定義を持つため、計算する際にいちいちデータの型が乗算できるものかどうかをチェックしたりする必要がないからだろうか？

```python
y=[]
N=100
proc=subprocess.run(["gcc","-o","homework1c","homework1.c"],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
for i in range(1,N+1,1):
    proc = subprocess.run(["./homework1c",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8").split()[0]))
    #print(proc.stdout.decode("utf8"))
x=list(range(1,N+1,1))
plt.plot(x,y)
plt.show()
```
