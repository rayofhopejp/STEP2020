```python
import subprocess
import matplotlib.pyplot as plt
%matplotlib inline
N=100
y=[]
for i in range(1,N+1):
    proc = subprocess.run(["python3","homework1.py",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8")))
    #print(proc.stderr.decode("utf8"))
x=list(range(1,N+1))
plt.plot(x,y)
plt.show()
```

```python
y=[]
proc=subprocess.run(["gcc","-o","homework1","homework1.c"],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
for i in range(1,N+1):
    proc = subprocess.run(["./homework1",str(i)],stdout = subprocess.PIPE, stderr = subprocess.PIPE)
    y.append(float(proc.stdout.decode("utf8")))
    #print(proc.stderr.decode("utf8"))
x=list(range(1,N+1))
plt.plot(x,y)
plt.show()
```
